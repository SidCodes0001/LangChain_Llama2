```python
from langchain.llms import HuggingFacePipeline
from transformers import AutoTokenizer
import transformers
import torch
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain
import warnings
warnings.filterwarnings('ignore')
```

# **Loading the Llama 2 Model**

```python
model="meta-llama/Llama-2-7b-chat-hf" # make sure you have access in hugging face account to this model
```

```python
tokenizer = AutoTokenizer.from_pretrained(
    model,
    padding=True,
    truncation=True,
    )
```

```python
pipeline = transformers.pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    torch_dtype=torch.bfloat16,
    trust_remote_code=True,
    device_map="auto",
    max_length=1000,
    do_sample=True,
    top_k=10,
    num_return_sequences=1,
    eos_token_id=tokenizer.eos_token_id
    )
```

```python
llm=HuggingFacePipeline(pipeline=pipeline, model_kwargs={'temperature':0})


```

```python
prompt="What would be a good name for a company that makes colorful socks"
print(llm(prompt))
```

# Prompt Templates

In many Large Language Model applications we donot pass the user input directly to the Large Language Model, we add the user input to a large piece of text called prompt template



## example

```python
prompt_template1=PromptTemplate(
    input_variables=["cuisine"],
    template="I want to open a restaurant for {cuisine} food. Suggest a fency name for this"
    )
```

```python
input_prompt=prompt_template1.format(cuisine="indian")
print(input_prompt)
```

## example 2


```python
prompt_template2=PromptTemplate(
    input_variables=["book_name"],
    template="Provide me a concise summary of the book {book_name}")
```

```python
input_prompt=prompt_template2.format(book_name="Alchemist")
print(input_prompt)
```

```python
chain = LLMChain(llm=llm, prompt=prompt_template2, verbose=True)
response= chain.run("Harry Potter")
print(response)
```

```python
chain2 = llm | prompt_template2
response= chain2.invoke("Harry Potter")
print(response)
```
